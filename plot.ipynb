```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the precision and recall values from the provided results
precision = {
    1: [100, 100, 100],  # Precision for nret = 1 (Query 1, Query 2, Query 3)
    2: [50, 50, 100],    # Precision for nret = 2
    3: [33.33, 33.33, 66.67]  # Precision for nret = 3
}

recall = {
    1: [100, 100, 50],  # Recall for nret = 1
    2: [100, 100, 100],  # Recall for nret = 2
    3: [100, 100, 100]  # Recall for nret = 3
}

# Calculate the average precision and recall for each nret
avg_precision = [np.mean(precision[nret]) for nret in precision]
avg_recall = [np.mean(recall[nret]) for nret in recall]

# Plotting average precision with discrete x-axis
plt.figure(figsize=(10, 6))
plt.plot(precision.keys(), avg_precision, label='Average Precision', marker='o', linestyle='-', markersize=6)

plt.xlabel('Number of Documents Retrieved (nret)')
plt.ylabel('Average Precision (%)')
plt.title('Average Precision vs. Number of Documents Retrieved')
plt.xticks([1, 2, 3])  # Set discrete x-ticks to be 1, 2, 3
plt.legend(loc='lower left')
plt.show()

# Plotting average recall with discrete x-axis
plt.figure(figsize=(10, 6))
plt.plot(recall.keys(), avg_recall, label='Average Recall', marker='o', linestyle='-', markersize=6)

plt.xlabel('Number of Documents Retrieved (nret)')
plt.ylabel('Average Recall (%)')
plt.title('Average Recall vs. Number of Documents Retrieved')
plt.xticks([1, 2, 3])  # Set discrete x-ticks to be 1, 2, 3
plt.legend(loc='lower left')
plt.show()
```
